```python
import pandas as pd
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
```

```python
#df = pd.read_csv(r'C:\Users\braisc\OneDrive - Leidos-LeidosCorpUS\Documents\Python Scripts\Datasets\EPLShootingStats_20-21.csv')
df = pd.read_csv('EPLShootingStats.csv')
df['Nation'] = df['Nation'].str.split(' ',expand=True)[1]
df = df[df['90s'] > 0]              #remove players who haven't played a single game
df = df[df['Pos'] == 'FW']          #select forwards
df = df.fillna(0)                   #replace NaN with zeros
df.head(15)
```

```python
#get statistics
df_stats = df.describe()
df_stats = df_stats.drop(['Born', '90s', 'SoT', 'FK', 'PK', 'PKatt'],axis=1)
df_mean = df_stats[df_stats.index == 'mean']

#write mean to a list to plot later
c_values = df_mean.values.tolist()
```

```python
#get players
print("Enter player 1: ")
player_1 = input()
print(player_1)

print("Enter player 2: ")
player_2 = input()
print(player_2)
```

```python
#create df subset an
df_1 = df[df['Player'] == player_1]
df_2 = df[df['Player'] == player_2]

df = pd.concat([df_1,df_2]).reset_index()
df_copy = df
df = pd.concat([df,df_stats[df_stats.index == 'mean'], 
               df_stats[df_stats.index == 'min'], df_stats[df_stats.index == 'max']])
df
```

```python
#clean up df by removing uneccesary columns (data and strings)
df = df.drop(['index', 'Unnamed: 0', 'Nation', 'Pos', 'Squad', 'Age', 'Born', '90s', 'SoT', 'FK', 'PK', 'PKatt', 'Matches', '-9999\\'],axis=1)
df
```

```python
#get parameters
params = list(df.columns)
params = params[1:]
params
```

```python
#add values to lst of tuple pairs
ranges = []
a_values = []
b_values = []

for x in params:
    a = min(df[params][x])
    #a = a - (a * 0.25)

    b = max(df[params][x])
    #b = b + (b * 0.25)

    #c = max(df[params][x])
    #c = c + (c * 0.25)
    
    ranges.append((a, b))

for x in range(len(df['Player'])):
    if df['Player'][x] == player_1:
        a_values = df.iloc[x].values.tolist()
    if df['Player'][x] == player_2:
        b_values = df.iloc[x].values.tolist()

a_values = a_values[1:]
b_values = b_values[1:]
#c_values = c_values[1:]

values = [a_values,b_values]
```

```python
values
```

```python
player_1_team = df_copy.loc[0,'Squad']
player_2_team = df_copy.loc[1,'Squad']

player_1_games = str(round(df_copy.loc[0,'90s']))
player_2_games = str(round(df_copy.loc[1,'90s']))

title = dict(
    title_name = player_1,
    title_color = 'red',
    subtitle_name = f'{player_1_team} - {player_1_games} matches',
    subtitle_color = 'red',
    title_name_2 = player_2,
    title_color_2 = 'blue',
    subtitle_name_2 = f'{player_2_team} - {player_2_games} matches',
    subtitle_color_2 = 'blue',
    title_fontsize = 18,
    subtitle_fontsize = 15,
)

endnote = 'Christopher Brais - Data from Statsbomb/FBRef'
```

```python
radar = Radar()

fig,ax = radar.plot_radar(ranges=ranges, params=params,values=values,radar_color=['red', 'blue'], alphas=[0.75,0.25], title=title,endnote=endnote, compare=True)
```
